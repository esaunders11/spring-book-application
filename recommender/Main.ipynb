```python
import pandas as pd
```

```python
books = pd.read_csv('books.csv')
```

```python
books.head()
```

```python
books.describe()
```

```python
books.isnull().sum()
```

```python
books.columns
```

```python
books = books[['authors', 'title', 'genre', 'description', 'rating']]
```

```python
books['tags'] = books['description'] + books['genre']
```

```python
books
```

```python
new_data = books.drop(columns=['description', 'genre'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=6810, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "Spider's Web"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[1])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(book):
    index = new_data[new_data['title'] ==  book].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("Assassin's Apprentice")
```
